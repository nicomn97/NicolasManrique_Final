# Los programadores recomiendan leguajes de programación

Con la información recolectada por Kaggle en su encuesta anual a los usuarios de su plataforma, y limitando el análisis de dicho conjunto a las preguntas 1 (género), 2 (edad), 3 (pais de residencia), 4 (educación), 5 (título de pregrado), 8 (años de experiencia programando), 9 (compensación anual), 17 (lenguaje de programación predilecto), 18 (lenguaje de programación que recomendarían a un principiante) y 48 (si consideran que los algoritmos de ML son como cajas negras o no), se presenta información relativa a los lenguajes de programación que recomiendan los programadores, y se buscan relaciones entre el perfil de cada encuestado y el lenguaje que recomiendan.

```python
df1=pd.DataFrame()
df2=pd.DataFrame()
df3=pd.DataFrame()
df4=pd.DataFrame()
df5=pd.DataFrame()
df8=pd.DataFrame()
df9=pd.DataFrame()
df17=pd.DataFrame()
df48=pd.DataFrame()
```

```python
dataFrames=[df1,df2,df3,df4,df5,df8,df9,df17,df48]
preg=['Q1','Q2','Q3','Q4','Q5','Q8','Q9','Q17','Q48']
k=0
for j in dataFrames:
    j['Recomendation'] = pd.Series(data['Q18'].unique())
    for i in data[preg[k]].value_counts().index:
        j[i] = pd.Series()    
    x = 0
    for progs in j['Recomendation']:  
        result = data.loc[data['Q18'] == progs][preg[k]].value_counts()
        for index in result.index:
            j.loc[j[index].index == x, index] = result[index]
        x = x + 1
    j=j[j.Recomendation != 0]
    k=k+1
```

```python
train_data, test_data = sklearn.model_selection.train_test_split(data[['Q18','Q1','Q2','Q3','Q4','Q5','Q8','Q9','Q17','Q48']], train_size=0.7)
```

```python
pr_area = 1.0
best_pr_area = 0.0
n_estimators = 100

while n_estimators<200 and (pr_area>=best_pr_area):
    n_splits = 6
    kfold = sklearn.model_selection.KFold(n_splits=n_splits)
    areas = np.zeros(n_splits)
    model_forest = sklearn.ensemble.RandomForestClassifier(n_estimators=n_estimators)
    i = 0 
    for train_index, test_index in kfold.split(train_data):
        model_forest.fit(train_data[preg].iloc[train_index], train_data.Q17.iloc[train_index])
        probas = model_forest.predict_proba(train_data[preg].iloc[test_index])
        precision, recall, _ = sklearn.metrics.precision_recall_curve(train_data.Q17.iloc[test_index], probas[:,1])
        ii = np.argsort(recall)
        areas[i] = scipy.integrate.trapz(precision[ii], recall[ii])
        i+=1
        
    pr_area = areas.mean()
    print(n_estimators, pr_area, best_pr_area)
    if pr_area > best_pr_area:
        best_pr_area = pr_area
        n_estimators += 1
    
    
print(n_estimators)
```
